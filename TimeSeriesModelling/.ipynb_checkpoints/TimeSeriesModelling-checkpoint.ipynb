```python
def retrieve_dataset(dataset_dict):
    url = dataset_dict['url']
    time_column_name = dataset_dict['time_column_name']
    value_column_name = dataset_dict['value_column_name']
    
    df = pd.read_csv(url)

    times = df[time_column_name]
    values = df[value_column_name]

    dataset_dict['times'] = times
    dataset_dict['values'] = values

    return dataset_dict
    
```
